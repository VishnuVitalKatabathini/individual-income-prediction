```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# Copy data and encode categorical features
df_ml = df.copy()

# Encode target variable
df_ml['income'] = df_ml['income'].str.strip()  # remove leading/trailing spaces
df_ml['income'] = df_ml['income'].map({'<=50K': 0, '>50K': 1})


# Label encode categorical features
categorical_cols = df_ml.select_dtypes(include='object').columns
label_encoders = {}
for col in categorical_cols:
    le = LabelEncoder()
    df_ml[col] = le.fit_transform(df_ml[col])
    label_encoders[col] = le

# Split features and target
X = df_ml.drop('income', axis=1)
y = df_ml['income']

# Standardize numerical features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Train Random Forest Classifier
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

# Predict and evaluate
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)


```

```python
print('accuracy :',accuracy)
print('classification report:\n',report)
print('confusion matrix:\n',conf_matrix)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score
import joblib

# Load training data
df = df.copy()

# Clean whitespaces
df = df.map(lambda x: x.strip() if isinstance(x, str) else x)

# Drop unwanted column
if "Unnamed: 0" in df.columns:
    df.drop(columns=["Unnamed: 0"], inplace=True)

# Remove rows with invalid 'age'
df = df[pd.to_numeric(df['age'], errors='coerce').notnull()]

# Encode target
df['income'] = df['income'].str.replace(".", "", regex=False).map({'<=50K': 0, '>50K': 1})

# Fill missing values
for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = df[col].fillna(df[col].mode()[0])
    else:
        df[col] = df[col].fillna(df[col].median())

# Label encode categorical columns
label_encoders = {}
for col in df.select_dtypes(include='object').columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le

# Split into features and labels
X = df.drop("income", axis=1)
y = df["income"]

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train-test split
X_train, X_val, y_train, y_val = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Train Logistic Regression
model = LogisticRegression(max_iter=1000, solver='lbfgs')  # Increase max_iter if needed
model.fit(X_train, y_train)

# Evaluate model
y_pred = model.predict(X_val)
print("Accuracy:", accuracy_score(y_val, y_pred))
print("\n Classification Report:\n", classification_report(y_val, y_pred))

# Save model and tools
joblib.dump(model, "logreg_income_model.joblib")
joblib.dump(label_encoders, "logreg_label_encoders.joblib")
joblib.dump(scaler, "logreg_scaler.joblib")
```

```python
# Re-importing necessary libraries after code execution state reset
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, roc_curve, auc
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_classification

# Creating sample dataset for demonstration (replace with your own X_train, X_test, y_train, y_test)
X, y = make_classification(n_samples=1000, n_features=20, n_informative=10, n_classes=2, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train Logistic Regression
log_model = LogisticRegression()
log_model.fit(X_train, y_train)
log_preds = log_model.predict(X_test)

# Train Random Forest
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)
rf_preds = rf_model.predict(X_test)

# Compute ROC curves
log_fpr, log_tpr, _ = roc_curve(y_test, log_model.predict_proba(X_test)[:, 1])
rf_fpr, rf_tpr, _ = roc_curve(y_test, rf_model.predict_proba(X_test)[:, 1])

# Create plots
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Confusion Matrix - Logistic Regression
sns.heatmap(confusion_matrix(y_test, log_preds), annot=True, fmt='d', cmap='Blues', ax=axes[0])
axes[0].set_title("Logistic Regression\nConfusion Matrix")
axes[0].set_xlabel("Predicted")
axes[0].set_ylabel("Actual")

# Confusion Matrix - Random Forest
sns.heatmap(confusion_matrix(y_test, rf_preds), annot=True, fmt='d', cmap='Greens', ax=axes[1])
axes[1].set_title("Random Forest\nConfusion Matrix")
axes[1].set_xlabel("Predicted")
axes[1].set_ylabel("Actual")

# ROC Curve
axes[2].plot(log_fpr, log_tpr, label=f"Logistic Regression (AUC = {auc(log_fpr, log_tpr):.2f})")
axes[2].plot(rf_fpr, rf_tpr, label=f"Random Forest (AUC = {auc(rf_fpr, rf_tpr):.2f})")
axes[2].plot([0, 1], [0, 1], 'k--')
axes[2].set_title("ROC Curve Comparison")
axes[2].set_xlabel("False Positive Rate")
axes[2].set_ylabel("True Positive Rate")
axes[2].legend()
axes[2].grid(True)

plt.tight_layout()
plt.show()
```
